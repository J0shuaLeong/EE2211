```python
import pandas as pd
import matplotlib.pyplot as plt
import csv

x = [] 
y = []

with open('GovernmentExpenditureOnEducation.csv','r') as csvfile: 
    lines = csv.reader(csvfile, delimiter=',') 
    for row in lines: 
        x.append(row[0]) 
        y.append(row[1]) 
        
plt.plot(x, y) 

plt.xlabel('Year') 
plt.ylabel('Expenditure ($)') 
plt.title('Gov Expenditure', fontsize = 20) 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('AnnualMotorVehiclePopulationbyVehicleType.csv')

# Filter the DataFrame for Omnibuses, Excursion buses, and Private buses
filtered_df = df[df['type'].isin(['Omnibuses', 'Excursion buses', 'Private buses'])]

# Pivot the data to have separate columns for each vehicle type
pivoted_df = filtered_df.pivot_table(index='year', columns='type', values='number', aggfunc='sum')

# Plot the line graph for each vehicle type
lines = pivoted_df.plot(linestyle='-')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number of Vehicles')
plt.title('Number of vehicles over the years')

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.legend(title='Vehicle Type')
plt.show()
```

```python
from sklearn.datasets import load_iris
Dfrom pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()

# Convert the data into a DataFrame
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the target variable (species) to the DataFrame
iris_df['species'] = iris.target

# Map target variable to species names
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Create scatter matrix
scatter_matrix(iris_df, figsize = (10, 10), c = iris.target, marker = 'o', hist_kwds = {'bins': 20}, s = 60, alpha = 0.8)

# Add labels and title
plt.show()
```
