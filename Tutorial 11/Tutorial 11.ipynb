```python
## Import necessary libraries
import random as rd
import numpy as np # linear algebra
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

## Generate data
## Set three centers, the model should predict similar results
center_1 = np.array([2,2])
center_2 = np.array([4,4])
center_3 = np.array([6,1])

## Generate random data and center it to the three centers
data_1 = np.random.randn(200, 2) + center_1
data_2 = np.random.randn(200,2) + center_2
data_3 = np.random.randn(200,2) + center_3
data = np.concatenate((data_1, data_2, data_3), axis = 0)

kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
centers = kmeans.cluster_centers_
labels = kmeans.labels_

#plt.scatter(data[:,0], data[:,1], s=7)
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', alpha=0.5, edgecolors='k')
plt.scatter(centers[:, 0], centers[:, 1], marker='x', c='red', s=200, label='Centroids')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('K-means Clustering')
plt.legend()
plt.show()
```

## Question 7

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score, homogeneity_completeness_v_measure

iris = load_iris()
X, y = iris.data, iris.target

k = 3
kmeans = KMeans(n_clusters=k)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
y_pred = kmeans.fit_predict(X)

plt.figure(figsize=(8, 6))

colors = ['blue', 'green', 'red']
for i in range(k):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], c=colors[i], label=f'Cluster {i+1}', alpha=0.7)

plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', c='black', s=200, label='Centroids')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('K-means Clustering on Iris Dataset')
plt.legend()
plt.grid(True)
plt.show()

ari = adjusted_rand_score(y, y_pred)
ami = adjusted_mutual_info_score(y, y_pred)
homo, compl, v_meas = homogeneity_completeness_v_measure(y, y_pred)

# Print the clustering metrics
print(f"Adjusted Rand Index (ARI): {ari}")
print(f"Adjusted Mutual Information (AMI): {ami}")
print(f"Homogeneity: {homo}")
print(f"Completeness: {compl}")
print(f"V-measure: {v_meas}")
```
