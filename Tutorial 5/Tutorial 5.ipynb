```python
# Question 6

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.metrics as metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


wine = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv",sep=';')

# Input Feature
X = wine.drop('quality', axis = 1)
# Target Output 
y = wine['quality']

# X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=1500, test_size=99, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=99/1599, random_state=42)

# train_X = X[0:1500]
# train_y = y[0:1500]
# test_X = X[1500:1599]
# test_y = y[1500:1599]

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)

# Perform linear regression on the training set
model = LinearRegression()
model.fit(X_train, y_train)
test_prediction = model.predict(X_test)
mse = metrics.mean_squared_error(y_test, test_prediction)


# Print out the learned parameters (coefficients and intercept)
print("Learned coefficients: ", model.coef_)
print("Learned intercept: ", model.intercept_)
print("MSE: ", mse)
```

```python
# Question 5

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('GovernmentExpenditureonEducation.csv')

X = data.drop('total_expenditure_on_education', axis = 1)
y = data['total_expenditure_on_education']

model = LinearRegression()
model.fit(X, y)

year_2021 = np.array([[2021]])
predicted_expenditure_2021 = model.predict(year_2021)
print("Predicted educational expenditure for the year 2021:", predicted_expenditure_2021[0])

plt.figure(figsize = (10, 6))
plt.scatter(X, y, color = 'blue', label = 'Actual')
plt.plot(X, model.predict(X), color = 'red', label = 'Predicted')
plt.title('Government Expenditure on Education Over the Years')
plt.xlabel('Year')
plt.ylabel('Educational Expenditure (in millions)')
plt.legend()
plt.grid(True)
plt.show()
```
